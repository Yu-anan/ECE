Problem 4

```python
x=[]
y=[]
for i in range(1,12):
    k=0.95+(i-1)/100
    x.append(k)
    q=1.95+(i-1)/100
    y.append(q)
for i in range(11):
    for j in range (11):
        plt.scatter(x[i],y[j])
plt.show()
w=[]
v=[]
for i in range(11):
    for j in range(11):
        w.append(x[i]*x[i]+5*y[j]*y[j])
        v.append(-5*x[i]*x[i]+y[j]*y[j])

plt.scatter(w,v)
plt.show()
     
sq1=(x[10]-x[0])*(y[10]-y[0])
l=np.sqrt((w[10]-w[0])*(w[10]-w[0])+(v[10]-v[0])*(v[10]-v[0]))
l2=np.sqrt((w[120]-w[10])*(w[120]-w[10])+(v[120]-v[10])*(v[120]-v[10]))
ans=l*l2
print('the square1 is',sq1)
print('the square2 is',ans)
print('ratio is' ,ans/sq1)
```

Problem 6

```python
mu=1
sigma=4
N=500
temp1=[]
temp2=[]
temp3=[]
min=0
max=2
for i in range(1,N):
    data=np.random.normal(mu,sigma,i)
    temp1.append(np.mean(data))
    data2=np.random.uniform(min,max,i)
    temp2.append(np.mean(data2))
    
for i in range(1,N):
    temp3.append(i)
plt.plot(temp3,temp1,label='Norm')
plt.plot(temp3,temp2,label='Uni')
plt.legend()
plt.show()
```
