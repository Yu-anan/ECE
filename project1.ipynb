```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
import pandas as pd
import math
list={100,1000,10000,100000}
l=0.5 #length og needle
x=[]
y=[]
def stimulate(n):
    data=np.random.uniform(0,0.5,n)
    data2=np.random.uniform(0,np.pi*0.5,n)
    k=0# intersection point
    for i in range(len(data)):
        if (data[i]<0.5*np.sin(data2[i])):
            x.append(data[i])
            y.append(data2[i])
            k=k+1
    pi=2*n/k
    print('pi is',pi)
    plt.figure()    # 定义一个图像窗口
    plt.plot(y,x) # 绘制曲线 y1
    plt.show()
    return pi
  
```

```python
for n in list:
    stimulate(n)
```

```python
import math
def nCr(n,r):
    b=0
    for i in range(n-r+1,n+1):
        b=b+math.log(i)
    q=0
    for k in range(1,r+1):
        q=q+math.log(k)
    answer=math.exp(b-q)
    return answer 
result=[]
n=[95,96,97,98,99,100]


for k in n:
    P=nCr(940,k)*nCr(60,100-k)/nCr(1000,100)
    result.append(P)
    

print('result',result)    
```
